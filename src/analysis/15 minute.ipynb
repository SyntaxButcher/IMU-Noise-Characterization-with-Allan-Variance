```python
import csv
import matplotlib.pyplot as plt
import numpy as np
import sklearn
from sklearn.metrics import mean_squared_error
from mpl_toolkits.mplot3d import Axes3D
import statistics as stat
from tabulate import tabulate


file = open('/home/shak/catkin_ws/src/lab3/rosbag/IMU_data.csv') #for still data in open area


csvreader = csv.reader(file)

header = []
header = next(csvreader)

rows = []
for row in csvreader:
		rows.append(row)

time = []
first = rows[0][0]

for i in rows:
	timeInSeconds = int(float(i[0])) - int(float(first))
	time.append(timeInSeconds)	

OrientX = []
for i in rows:
	OrientX.append(float(i[12]))

OrientY = []
for i in rows:
	OrientY.append(float(i[13]))
	
OrientZ = []
for i in rows:
	OrientZ.append(float(i[14]))
	
OrientW = []
for i in rows:
	OrientW.append(float(i[15]))
	
GyroX = []
for i in rows:
	GyroX.append(float(i[17]))
	
GyroY = []
for i in rows:
	GyroY.append(float(i[18]))

GyroZ = []
for i in rows:
	GyroZ.append(float(i[19]))
	
LinX = []
for i in rows:
	LinX.append(float(i[21]))
	
LinY = []
for i in rows:
	LinY.append(float(i[22]))
	
LinZ = []
for i in rows:
	LinZ.append(float(i[23]))
	
MagX = []
for i in rows:
	MagX.append(float(i[29]))
	
MagY = []
for i in rows:
	MagY.append(float(i[30]))

MagZ = []
for i in rows:
	MagZ.append(float(i[31]))
	
```

```python
plt.scatter(time,GyroX, color = 'red', s = 10)
evn = plt.gca()
evn.set_xlabel('Time (seconds)', fontsize=10)
evn.set_ylabel('Angular Velocity(rad/s)', fontsize=10)
evn.set_title('Angular Velocity(around X) vs Time', fontsize=15)
plt.show()
```

```python
plt.scatter(time,GyroY, color = 'blue', s = 10)
evn = plt.gca()
evn.set_xlabel('Time (seconds)', fontsize=10)
evn.set_ylabel('Angular Velocity(rad/s)', fontsize=10)
evn.set_title('Angular Velocity(around Y) vs Time', fontsize=15)
plt.show()
```

```python
plt.scatter(time,GyroZ, color = 'green', s = 10)
evn = plt.gca()
evn.set_xlabel('Time (seconds)', fontsize=10)
evn.set_ylabel('Angular Velocity(rad/s)', fontsize=10)
evn.set_title('Angular Velocity(around Z) vs Time', fontsize=15)
plt.show()
```

```python
#Mean and SD
MeanGyroX = stat.mean(GyroX)
MeanGyroY = stat.mean(GyroY)
MeanGyroZ = stat.mean(GyroZ)
StdGyroX = stat.stdev(GyroX)
StdGyroY = stat.stdev(GyroY)
StdGyroZ = stat.stdev(GyroZ)
TableGyro = [["Gyro Axis","Mean","Standard Deviation"],
             ["X",MeanGyroX,StdGyroX],
             ["Y",MeanGyroY,StdGyroY],
             ["Z",MeanGyroZ,StdGyroZ]]
print(tabulate(TableGyro, headers='firstrow', tablefmt='fancy_grid'))
```
